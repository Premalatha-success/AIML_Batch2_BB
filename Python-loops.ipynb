## while loops

```python
num=15
addition=0
i=1
while i<=num:
    addition=addition+i
    i=i+1
print("Addition is:",addition)
```

### For loop

```python
number=range(1,16)
addition=0
i=1
for i in number:
    addition=addition+i
    i=i+1
print("Addition is:",addition)
```
